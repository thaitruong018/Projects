# 2. Create a questions and answers bank

We want something like this:

- questions: contains all the questions that you have 
multiple_choices: all the multiple choices option to give to our players
- answers: contains all the answers 

**NOTE**: Some lines contain wrong answers (it's not an option from the multiple choices) -> you need to clean these 


````python
# Read file
import random
with open ('wwbm_qa.txt') as f:
    sequences = []
    for line in f:
        sequences.append(line.strip())
````

# üëè 3. Create the millionaire game

````python
# Get data consists of questions and answers:
questions = []
raw_answer = []
for i in sequences:
    a= i.split('[')[0]
    b = i.split('[')[1]
    questions.append(a)
    raw_answer.append(b)
# Split raw answer 
multiple_choices = []
correct_answer = []
for i in raw_answer:
    c= i.split(',')[-1]
    d= i.split(',')[:-1]
    multiple_choices.append(d)
    correct_answer.append(c)
````

````python
def random_question(x):
    random_generate = random.sample(range(len(x)),1)[0]
    Q= questions[random_generate]
    W = multiple_choices[random_generate]
    Z = correct_answer[random_generate]
    return Q, W, Z
````
